```python
df = pd.read_csv('/kaggle/input/data-science-jobs-salaries/Data Science Jobs Salaries.csv')
df.head()
```

```python
df.employee_residence.value_counts()
```

```python
us_emps = df[df['employee_residence'] == 'US']
us_emps.shape
```

```python
us_emps.info()
```

```python
us_emps['work_year'].replace({'2021e':'2021'}, inplace=True)
```

```python
us_emps.work_year.value_counts()
```

```python
us_emps.head()
```

```python
us_emps['work_year']=pd.to_datetime(us_emps['work_year'], format = '%Y')
```

```python
us_emps.info()
```

```python
us_emps['work_year']=us_emps.work_year.dt.year
```

```python
us_emps.head()
```

```python
us_emps.company_location.value_counts()
```

```python
us_emps['job_title'].value_counts()>=7
```

```python
# us_emps.loc[us_emps['job_title'].value_counts()>=7]
job_counts = us_emps['job_title'].value_counts().rename('job_counts')
job_counts

us_emps = us_emps.merge(job_counts.to_frame(),
                                left_on='job_title',
                                right_index=True)
```

```python
us_common_jobs = us_emps[us_emps.job_counts >=7]
us_common_jobs
```

```python
us_common_jobs.salary_currency.value_counts()
```

```python
us_common_jobs.drop(['salary_currency', 'employee_residence'], axis=1, inplace=True)
us_common_jobs.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```
